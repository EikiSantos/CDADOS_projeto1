#### Como a entrada no MERCOSUL impactou a economia do Brasil?



O Mercado Comum do Sul, ou Mercosul, é um bloco econômico criado em <font color="red" >1991 </font>, no tratado de Assunção, com a união dos países: Argentina, Brasil, Paraguai e Uruguai. A principal fonte de movimentação financeira do Mercosul são as exportações e importações. É importante ressaltar que o Brasil ocupa a posição de líder econômico do bloco, apresentando a maior população e o maior PIB dentre os países presentes no Mercosul.    

A Venezuela foi o ultimo país à entrar no Mercosul. Os países associados são: Bolívia, Colombia, Peru, Equador e Chile.

Usaremos dados da África do Sul para compara-los com os dados do Brasil, visto que ambos represtentam países com grande potencial de crescimento econômico, além de outras semelhanças.

### <font color = "#a29bec">Hipótese: </font>  
##### A admissão do Brasil no Mercosul aumentou o crescimento anual do PIB.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
baby = pd.read_excel('baby.xlsx')
gdp = pd.read_excel('gdp.xlsx')
hydro = pd.read_excel('hydro.xlsx')
imports = pd.read_excel("imports.xlsx")
idh = pd.read_excel("idh.xlsx")
inflation = pd.read_excel("inflation.xlsx")
importsgdp = pd.read_excel("imports_percent_of_gdp.xlsx")
exportsgdp = pd.read_excel("exports_percent_of_gdp.xlsx")
#------------------------------------------------------------------------------------------------------------------------------

nuclearp = pd.read_excel('np.xlsx')
hydrop = pd.read_excel('hp.xlsx')
gasp = pd.read_excel('ngp.xlsx')
```

```python
baby = baby.set_index("country")
gdp = gdp.set_index("country")
hydro = hydro.set_index("country")
imports = imports.set_index("country")
idh = idh.set_index("country")
inflation = inflation.set_index("country")
importsgdp = importsgdp.set_index("country")
exportsgdp = exportsgdp.set_index("country")
#------------------------------------------------------------------------------------------------------------------------------

nuclearp = nuclearp.set_index("country")
hydrop = hydrop.set_index('country')
gasp = gasp.set_index('country')
```

```python
baby = baby.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()
gdp = gdp.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()
hydro = hydro.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()
imports = imports.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()
idh = idh.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()
inflation = inflation.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()
importsgdp = importsgdp.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()
exportsgdp = exportsgdp.loc[["Brazil", "South Africa", "Austria", "Norway", "New Zealand", "Russia", "China", "India"]].transpose()

#------------------------------------------------------------------------------------------------------------------------------

nuclearp = nuclearp.loc[["Brazil", "South Africa"]].transpose()
hydrop = hydrop.loc[["Brazil", "South Africa"]].transpose()
gasp = gasp.loc[["Brazil", "South Africa"]].transpose()
```

```python
plt.plot(gdp.loc[(gdp.index > 1950) & (gdp.index < 2014), "Brazil"], label="Brazil", color = '#a29bec')
plt.plot(gdp.loc[(gdp.index > 1950) & (gdp.index < 2014), "South Africa"], label="South Africa", color = '#f7c066')
plt.grid(True)
plt.legend()
plt.title("GDP by Year growth (Brazil e South Africa)")
plt.show()
```

```python
plt.plot(gdp.loc[(gdp.index > 1990) & (gdp.index < 1996), "Brazil"], label="Brazil", color = '#a29bec')
plt.plot(gdp.loc[(gdp.index > 1990) & (gdp.index < 1996), "South Africa"], label="South Africa", color = '#f7c066')
plt.grid(True)
plt.legend()
plt.title("GDP by Year growth (Brazil e South Africa) (1990 - 1995)")
plt.show()
```

```python
gdp_br = []
i = 1
while i <= 5:
    gdpbr = gdp.loc[1990+i,"Brazil"]
    gdp_br.append(gdpbr)
    i+=1
print(gdp_br)
#--------------------------------------------------------------
gdp_sa = []
i = 1
while i <= 5:
    gdpsa = gdp.loc[1990+i,"South Africa"]
    gdp_sa.append(gdpsa)
    i+=1
print(gdp_sa)
#---------------------------------------------------------------
anos = [1991,1992,1993,1994,1995]
#---------------------------------------------------------------
#plt.bar(anos,gdp_br,color="#66B696")
#plt.ylabel("GDP growth by year")
#plt.xlabel("Ano")
#plt.title("Crescimento anual do PIB")
#plt.show
#---------------------------------------------------------------
#Largura das barras
barWidth = 0.25

#Aumentando o Gráfico
plt.figure(figsize=(10,5))

#Posições das barras
r1 = np.arange(len(gdp_br))
r2 = [x + barWidth for x in r1]

plt.bar(r1, gdp_br, color='#a29bec',width=barWidth,label='Brazil')
plt.bar(r2, gdp_sa, color="#f7c066",width=barWidth,label='South Africa')
plt.xlabel("Ano")
plt.xticks([r + barWidth for r in range(len(gdp_br))]), ['1991','1992','1993','1994','1995']
plt.ylabel("GDP")
plt.title("Crescimento Anual do PIB")
plt.legend()
plt.show()
```

É importante notar que o crescimento anual do Pib do Brasil a partir de 1991 cresceu muito mais em relação à África do Sul
